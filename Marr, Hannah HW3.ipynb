## Question 4: Volatility Analysis
### Objective: 
- Compute and analyze the rolling volatility of the ^DJI index over a specified window.

### Instructions:

- Calculate the rolling standard deviation for a specified window (e.g., 21 days) to estimate volatility.
- Plot and analyze periods of high and low volatility.

```python
# 4. Calculate the rolling standard deviation for a specified window (e.g., 21 days) to estimate volatility
# Calculate the 21-day rolling standard deviation to estimate volatility
rolling_volatility = daily_returns_cleaned.rolling(window=21).std()
rolling_volatility # We would expect the first 21 days to be blank here
```

```python
# 4. Plot and analyze periods of high and low volatility
import matplotlib.pyplot as plt

# Define thresholds for high and low volatility
# Ensure that the mean and std are scalar values by using .item() or .values[0] to convert them
high_vol_threshold = (rolling_volatility.mean() + rolling_volatility.std()).item()  # Extract as scalar
low_vol_threshold = (rolling_volatility.mean() - rolling_volatility.std()).item()  # Extract as scalar

# Plot the rolling volatility
plt.figure(figsize=(12, 6))

# Plot the rolling volatility as a simple line
plt.plot(rolling_volatility.index, rolling_volatility.values, label='21-Day Rolling Volatility', color='blue')

# Add horizontal lines for the high and low volatility thresholds
plt.axhline(y=high_vol_threshold, color='red', linestyle='--', label='High Vol Threshold')
plt.axhline(y=low_vol_threshold, color='green', linestyle='--', label='Low Vol Threshold')

# Add title and labels
plt.title('Periods of High and Low Volatility (2021-2024)')
plt.xlabel('Date')
plt.ylabel('Volatility (Standard Deviation)')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Question 5: Cumulative Return Analysis
### Objective: 
- Analyze the cumulative return of the ^DJI index for the given time period.

### Instructions:

- Calculate the cumulative daily returns.
- Use Matplotlib to visualize the cumulative return over the given period.

```python
# 5. Calculate the cumulative daily returns
# Calculate daily returns (percentage change)
daily_returns = closing_prices.pct_change()

# Calculate cumulative returns
cumulative_returns = (1 + daily_returns).cumprod()

cumulative_returns
```

```python
# 5. Use Matplotlib to visualize the cumulative return over the given period
# Plot the cumulative daily returns
plt.figure(figsize=(12, 6))
plt.plot(closing_prices.index, cumulative_returns, label='Cumulative Daily Returns', color='purple')

# Add title and labels
plt.title('Cumulative Daily Returns (2021-2024)')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

## Question 7: Historical High and Low Price Analysis
### Objective: 
- Analyze periods where the ^DJI index achieved historical highs and lows.

### Instructions:

- Identify dates where the ^DJI index achieved historical highs and lows.
- Use Matplotlib to visualize these points on the price chart.

```python
# 7. Identify dates where the ^DJI index achieved historical highs and lows
# Calculate cumulative historical highs and lows
data['Historical_High'] = data['Close'].cummax()
data['Historical_Low'] = data['Close'].cummin()

# Find the latest historical high and low
latest_high = data['Historical_High'].iloc[-1]
latest_low = data['Historical_Low'].iloc[-1]

# Find the dates when these historical highs and lows were first set
date_of_high = data['Historical_High'].idxmax()
date_of_low = data['Historical_Low'].idxmin()

# Output the results
print(f"The latest historical high of the ^DJI index is {latest_high} on {date_of_high}")
print(f"The latest historical low of the ^DJI index is {latest_low} on {date_of_low}")
```

```python
# 7. Use Matplotlib to visualize these points on the price chart
import matplotlib.pyplot as plt

# Plot the Close prices and historical highs and lows
plt.figure(figsize=(12, 6))

# Plot the Close prices
plt.plot(data.index, data['Close'], label='Close Prices', color='blue')

# Plot the Historical Highs and Lows
plt.plot(data.index, data['Historical_High'], label='Historical Highs', color='green', linestyle='--')
plt.plot(data.index, data['Historical_Low'], label='Historical Lows', color='red', linestyle='--')

# Highlight the most recent historical high and low
plt.scatter(data['Historical_High'].idxmax(), data['Historical_High'].max(), color='green', label='Latest High', zorder=5)
plt.scatter(data['Historical_Low'].idxmin(), data['Historical_Low'].min(), color='red', label='Latest Low', zorder=5)

# Add title and labels
plt.title('^DJI Price Chart with Historical Highs and Lows')
plt.xlabel('Date')
plt.ylabel('Price')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

# Output the latest historical high and low values and the dates
print(f"The latest historical high of the ^DJI index is {data['Historical_High'].max()} on {data['Historical_High'].idxmax()}")
print(f"The latest historical low of the ^DJI index is {data['Historical_Low'].min()} on {data['Historical_Low'].idxmin()}")
```
