## 3.  Advanced Data Aggregation with Pandas

### Objective:

#### - To gain proficiency in using advanced data aggregation methods in Pandas, such as groupby and pivot_table.

### Instructions:

#### 1. Load Dataset:
#### - Load the Gapminder dataset. (https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv)
#### 2. Groupby Operations:
#### - Use groupby to find the average life expectancy and GDP per capita for each continent for the year 2007.
#### 3. Advanced Pivot Table:
#### - Create a pivot table with continent as an index, year as columns, and fill values with the mean of gdpPercap.
#### 4. Print Results (.head(5))

```python
# 1.
# Load the dataset
gapminder = pd.read_csv('https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv')

# Display the first 5 rows
gapminder.head(5)
```

```python
# 2.
# Filter the dataset for the year 2007
gapminder_2007 = gapminder[gapminder['year'] == 2007]

# Group by continent and calculate the average life expectancy and GDP per capita
continent_stats = gapminder_2007.groupby('continent').agg(
    avg_life_expectancy=('lifeExp', 'mean'),
    avg_gdp_per_capita=('gdpPercap', 'mean')
)

# Display the first 5 rows
continent_stats.head(5)
```

```python
# 3.
# Create a pivot table with continent as index, year as columns, and mean of gdpPercap as values
pivot_table = gapminder.pivot_table(
    values='gdpPercap', 
    index='continent', 
    columns='year', 
    aggfunc='mean'
)

# 4.
# Display the first 5 rows
pivot_table.head(5)
```

## 6. Comprehensive Data Analysis with Pandas

### Objective:

#### - To synthesize knowledge of Pandas, Numpy, data cleaning, manipulation, and aggregation to perform comprehensive analysis on a real-world dataset.

### Instructions:

#### 1. Load the Auto MPG Dataset (https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original):
#### - Download the Auto MPG dataset and load it into a DataFrame.
#### - Use appropriate column names based on the dataset documentation.
#### 2. Data Cleaning:
#### - Identify and drop any missing or anomalous values.
#### 3. Data Transformation and Feature Engineering:
#### - Compute the Displacement-to-Weight ratio and add it as a new feature in the DataFrame.
#### - Extract the car Brand from the car name column and create a new column, Brand.
#### 4. Data Aggregation:
#### - Group the dataset by Brand and model year and calculate the average mpg for each group.
#### - Create a pivot table to illustrate the average mpg of cars for each combination of Brand and model year.
#### 5. Advanced Filtering and Sorting:
#### - Identify the top 5 cars with the highest mpg within each model year.
#### - Create a new DataFrame containing only these top 5 cars per model year.
#### 6. Print Results (.head(5))

```python
# 1.
# Define the column names based on the dataset documentation
column_names = [
    'mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
    'acceleration', 'model_year', 'origin', 'car_name'
]

# Load the dataset and specify the delimiter
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original'
auto_mpg = pd.read_csv(url, delim_whitespace=True, names=column_names)

# Display the first 5 rows of the dataset
auto_mpg.head(5)
```

```python
# 2.
# Identify missing values
missing_values = auto_mpg.isna().sum()
print("Missing values in each column:")
print(missing_values)

# Drop rows with any missing values
auto_mpg_cleaned = auto_mpg.dropna()

# Display the first 5 rows of the cleaned dataset
auto_mpg_cleaned.head(5)
```

```python
# 3.
# Compute the Displacement-to-Weight ratio by dividing the displacement column by the weight column
auto_mpg_cleaned['displacement_to_weight'] = auto_mpg_cleaned['displacement'] / auto_mpg_cleaned['weight']

# Extract the car Brand from the 'car_name' column
# Assuming the Brand is the first word in the car_name, we'll split by space and take the first word
auto_mpg_cleaned['Brand'] = auto_mpg_cleaned['car_name'].str.split().str[0]

# Display the first 5 rows of the dataset with the new features
auto_mpg_cleaned.head(5)
```

```python
# 4.
# Group by Brand and model year, and calculate the average mpg
grouped_mpg = auto_mpg_cleaned.groupby(['Brand', 'model_year'])['mpg'].mean().reset_index()

# Create a pivot table with Brand as the index, model_year as columns, and average mpg as values
pivot_table_mpg = auto_mpg_cleaned.pivot_table(values='mpg', index='Brand', columns='model_year', aggfunc='mean')
```

```python
# Display first 5 rows of grouped data
grouped_mpg.head(5)
```

```python
# Display first 5 rows of pivot table
pivot_table_mpg.head(5)
# Leaving null values in; since I cleaned this data, NaN values correspond to brand/model year combinations that do not exist in this dataset
```

```python
# 5.
# Group the dataset by 'model_year' and find the top 5 cars with the highest mpg within each group
top_5_cars_per_year = auto_mpg_cleaned.groupby('model_year').apply(lambda x: x.nlargest(5, 'mpg')).reset_index(drop=True)

# 6.
# Display the first 5 rows of the new dataframe
top_5_cars_per_year.head(5)
```
